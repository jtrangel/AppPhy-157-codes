```python
import numpy as np
import matplotlib.pyplot as plt
from math import sin,pi
#general array set-up
N = 200 #N x N pixels
x = np.linspace(-2,2,N)
y = x
x,y = np.meshgrid(x,y)
A = np.zeros(np.shape(x))

#Sine wave with amplitude from -1 to 1
for n in range(N):
  A[:,n] = 2*(((sin(2*pi*4*n/N))**2) - 0.5)

plt.imshow(A, cmap = "gray")

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(x,y,A)

Aim = A*255
img = Aim.astype(np.uint8)
plt.imsave("1.1.jpg",img)
plt.imsave("1.1.bmp",img)
plt.imsave("1.1.png",img)
plt.imsave("1.1.tiff",img)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from math import sin,pi
N = 200 #N x N pixels
x = np.linspace(-2,2,N)
y = x
x,y = np.meshgrid(x,y)
A = np.zeros(np.shape(x))

#Formation of square aperture
for m in range(N):
  for n in range(N):
    if x[m,n]>-1 and x[m,n]<1:
      if y[m,n]>-1 and y[m,n]<1:
        A[m,n] = 1.0
      else: 
        A[m,n] = 0.0

plt.imshow(A, cmap = "gray")

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(x,y,A)

Aim = A*255
img = Aim.astype(np.uint8)
plt.imsave("1.3.jpg",img)
plt.imsave("1.3.bmp",img)
plt.imsave("1.3.png",img)
plt.imsave("1.3.tiff",img)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from math import sin,pi
N = 200 #N x N pixels
x = np.linspace(-2,2,N)
y = x
x,y = np.meshgrid(x,y)
R = np.sqrt(x**2 + y**2)
A = np.zeros(np.shape(x))

#Formation of annular ring
for m in range(N):
  for n in range(N):
    if R[m,n]>1.5 and R[m,n]<2:
      A[m,n] = 1.0
    else: 
      A[m,n] = 0.0

plt.imshow(A, cmap = "gray")

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(x,y,A)

Aim = A*255
img = Aim.astype(np.uint8)
plt.imsave("1.4.jpg",img)
plt.imsave("1.4.bmp",img)
plt.imsave("1.4.png",img)
plt.imsave("1.4.tiff",img)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from math import sin,pi,exp
from scipy.stats import multivariate_normal

N = 200 #N x N pixels
x = np.linspace(-2,2,N)
y = x
x,y = np.meshgrid(x,y)
A = np.zeros(np.shape(x))

#Multivariate gaussian using scipy.stats package
xy = np.column_stack([x.flat, y.flat])
mu = np.array([0.0, 0.0])
sigma = np.array([0.72, 0.72])
covariance = np.diag(sigma**2)
A = multivariate_normal.pdf(xy, mean= mu, cov=covariance)
A = A.reshape(x.shape)

#Indexes to approximate R=1.75 cm
indexx = A[0]
print(indexx[12])
print(indexx[188])
plt.imshow(A, cmap = "gray")

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(x,y,A)

Aim = A*255
img = Aim.astype(np.uint8)
plt.imsave("1.5.jpg",img)
plt.imsave("1.5.bmp",img)
plt.imsave("1.5.png",img)
plt.imsave("1.5.tiff",img)
```

```python
import numpy as np
import matplotlib.pyplot as plt
#set-up 
N = 500
x = np.linspace(-10,10,num = N)
y = x
X,Y = np.meshgrid(x,y)
Rd, Gn, Bl = np.zeros((N,N)),np.zeros((N,N)), np.zeros((N,N)) 

#White background
Rd[...] = 1.0
Gn[...] = 1.0
Bl[...] = 1.0

#draw colored circles
Rt, Rc, deg = 3, 3, 30
xt, yt = Rt*np.cos(deg*np.pi/180) + 0.7,Rt*np.sin(deg*np.pi/180) - 1.2


R = np.sqrt((X+2*xt)**2 + (Y+Rt)**2)
for m in range(N):
  for n in range(N):
    if R[m,n]>2.55 and R[m,n]<Rc:
      Rd[m,n] = 0.0
      Gn[m,n] = 0.0
      Bl[m,n] = 1.0 #Blue circle
R = np.sqrt((X-2*xt)**2 + (Y+Rt)**2)
for m in range(N):
  for n in range(N):
    if R[m,n]>2.55 and R[m,n]<Rc:
      Rd[m,n] = 1.0 #Red Circle
      Gn[m,n] = 0.0
      Bl[m,n] = 0.0 
R = np.sqrt((X)**2 + (Y+Rt)**2)
for m in range(N):
  for n in range(N):
    if R[m,n]>2.55 and R[m,n]<Rc:
      Rd[m,n] = 0.0
      Gn[m,n] = 0.0 #Black circle
      Bl[m,n] = 0.0 
R = np.sqrt((X-xt)**2 + (Y-yt)**2)
for m in range(N):
  for n in range(N):
    if R[m,n]>2.55 and R[m,n]<Rc:
      Rd[m,n] = 0.0
      Gn[m,n] = 1.0 #Green Circle
      Bl[m,n] = 0.0
R = np.sqrt((X+xt)**2 + (Y-yt)**2)
for m in range(N):
  for n in range(N):
    if R[m,n]>2.55 and R[m,n]<Rc:
      Rd[m,n] = 1.0
      Gn[m,n] = 1.0 #Yellow circle: red+green
      Bl[m,n] = 0.0
I = np.zeros((N,N,3))
I[...,0] = Rd
I[...,1] = Gn
I[...,2] = Bl
fig = plt.figure()
plt.imshow(I)

Iim = I*255
img = Iim.astype(np.uint8)
plt.imsave("1.6.jpg",img)
plt.imsave("1.6.bmp",img)
plt.imsave("1.6.png",img)
plt.imsave("1.6.tiff",img)
```

```python
#Contrast stretching
def Constretch(color,perc):
  x = np.arange(0,256)
  colorsh = np.reshape(color,(-1,1)) #Turn the image array into a column vector
  PDF = plt.hist(colorsh,bins=x, density=True)
  Iold = np.cumsum(PDF[0])
  
  Cnew = colorsh[:,0]
  Inew = np.zeros(255)

  for n in range(len(colorsh)):  #Rounding of original image values
    Cnew[n]= round(Cnew[n])

  #Imin and Imax
  Imin = Iold[np.where(Iold[...]>perc)][0]   
  Imax = Iold[np.where(Iold[...]>(1-perc))][0]

  x1 = np.arange(0,255)
  for n in x1:           #Using the formula, rescaled by a factor of 255
    Inew[n] = (Iold[n] - Imin)*255/(Imax-Imin) 
    if Inew[n] <0: 
      Inew[n] = 0        #Setting byte limit
    if Inew[n] >255:
      Inew[n] = 255

  for n in x1:           #Adjusting greyscale value of each pixel
    Cnew[np.where(Cnew[...] == n)] = Inew[n]
  Contrasted = np.reshape(Cnew,np.shape(color))
  return Contrasted
```

```python
#Taking averages
def avg(color):
  shape = np.shape(color) #take shape of array
  sum = 0.0
  total = shape[0]*shape[1] #total pixels

  #taking total sum
  for m in range(shape[0]):
    for n in range(shape[1]): 
      sum += color[m,n] 
  average = sum/total 
  return average
```

```python
Rave = avg(Rc)
Gave = avg(Gc)
Bave = avg(Bc)
Rwb = Rc/Rave
Gwb = Gc/Gave
Bwb = Bc/Bave
```
